### Descrição Matemática

---
<UL>
<LI><p align="justify">Uma onda senoidal é um padrão que descreve um aumento gradual e uma subsequente diminuição repetitiva de uma quantidade ao longo do tempo. Essa onda é fundamental para entender sistemas elétricos, especialmente as variações de tensão e corrente alternada (CA).</p></LI>

<LI><p align="justify">Tanto a tensão quanto a corrente em sistemas de energia elétrica são representadas através de ondas senoidais. A mudança da perspectiva do domínio do tempo para o domínio da frequência ajuda a analisar e projetar essas ondas com maior eficiência.</p></LI>

<LI><p align="justify">Uma função senoidal é caracterizada por três componentes principais: amplitude, frequência e fase. A amplitude é o ponto mais alto da onda, medido a partir da posição neutra. A frequência define quantos ciclos completos ocorrem em uma unidade de tempo. A fase determina o ponto de partida da onda senoidal no eixo do tempo.</p></LI>

<LI><p align="justify">A amplitude é a distância do ponto zero até o pico da onda e também a metade do intervalo total entre pico e vale. A frequência se refere ao ritmo de oscilação da onda, enquanto o período é o inverso da frequência e denota a duração de um ciclo completo. A fase indica quão deslocada a onda está em relação a um ponto de referência no tempo zero.</p></LI>

<LI><p align="justify">Expressar o tempo como um ângulo é uma maneira eficaz de lidar com funções senoidais. O argumento de uma função senoidal precisa ser adimensional, o que significa que não tem uma unidade de medida física. Um ângulo, mesmo medido em graus ou radianos, não carrega uma dimensão física intrínseca, mas sim representa uma proporção ou fração de um ciclo.</p></LI>

<LI><p align="justify">Graficamente, uma função senoidal é plotada em relação ao ângulo no eixo horizontal. A altura da curva em qualquer ponto é determinada pelo valor do seno desse ângulo, multiplicado pela amplitude. A função senoidal completa um ciclo quando o ângulo aumenta até 360 graus, repetindo seu padrão.</p></LI>

<LI><p align="justify">Ao trabalhar com funções senoidais, é comum usar radianos em vez de graus para medir ângulos. A conversão é simples: 360 graus equivalem a 2π radianos. Radianos são preferíveis porque são uma medida mais natural quando lidamos com círculos e oscilações. A conversão de graus para radianos pode ser expressa pela  seguinte fórmula:</p></LI>

$$
\frac{360}{ângulo} = \frac{2\pi}{x} \rightarrow x = \frac{2\pi \cdot ângulo}{360} radianos
$$

<LI><p align="justify">A frequência angular $\omega$ expressa a taxa de mudança de ângulo em relação ao tempo. Enquanto a frequência $f$ mede o número de ciclos completos em um segundo. A frequência angular corresponde a 2π vezes a frequência em hertz (Hz).</p></LI>

<LI><p align="justify">Uma frequência angular de 60 ciclos por segundo (60 Hz) equivale a aproximadamente 377 radianos por segundo. Isso é relevante para a análise de sistemas elétricos, especialmente quando se trata de geração, transmissão e consumo de energia elétrica.</p></LI>

</UL>
















```python
%%manim -qm -v WARNING Senoide

class Senoide(Scene):
    def construct(self):
        self.camera.background_color=GRAY_E
        plane = NumberPlane(
            x_range = (0,2*PI),
            y_range = (-3,3),
            axis_config={
                "include_ticks": None,
                "include_tip": None,
            },
        )
        plane.add_coordinates().scale(0.8)
        curve = plane.plot(lambda x: 2 * np.sin(x), color=RED)

        Grafico = VGroup(plane, curve).shift(RIGHT*2)

        label_center = Tex("Senoide:", color = ORANGE)
        label_center2 = Tex("A $\cdot$ sin($\omega$t + $\Phi$)")
        label_center2.next_to(label_center, RIGHT)

        box1 = Rectangle(height = 3, width = 5, color = PURPLE).next_to(curve, LEFT, buff = 1)

        label_box11 = (Tex("A - Amplitude"))
        label_box12 = Tex(r"$\omega$ - Frequência angular")
        label_box13 = Tex(r"$\phi$ - Ângulo de fase")

        labels_box1 = VGroup(label_box11, label_box12, label_box13).scale(0.7)

        labels_box1.arrange(direction=DOWN).move_to(box1)

        Labels = VGroup(label_center, label_center2).next_to(plane,UP)
        self.add(Grafico, Labels, box1, labels_box1)
```

Para a Senoide:

$$
 A \cdot sen(\omega \cdot t + \phi)
$$
<UL>

<LI><p align="justify">O valor de Pico é o valor máximo ou amplitude da corrente. Como o valor de um seno ou cosseno varia entre 1 e -1, a corrente real oscila entre A e -A. O tempo T é medido em segundos e, quando multiplicado pela frequência angular $\omega$, dá um número de radianos. No caso mais simples, não há deslocamento de fase, significando que F é zero e a curva começa em zero.</p>

 <LI><p align="justify">Assim, temos que para o deslocamento de fase, pode ser análogo a variação da frêquencia, pois, também não altera as outras caracteristicas da onda, como pode ser visto na imagem abaixo, onde insere o angulo de defasagem e você verá a forma de onda.



```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

angle = float(input('Escreva o ângulo $\Phi$ de desafagem: '))
rad = angle*2*pi/360

t = np.arange(0, 2*pi, 0.001)
sen1 = 3*np.sin(t)
sen2 = 3*np.sin(t+rad)

fig, senoide = plt.subplots()
senoide.plot(t, sen1, 'b', t, sen2)

senoide.set(xlabel='time (s)', ylabel='Amplitude',
       title= 'Senoide: $A \cdot sin(\omega \cdot t + %.1f^{o})$' %(angle))

senoide.grid()
```

### O valor $rms$

---
<UL>
<LI><p align="justify"> Para a maioria das aplicações, a magnitude geral das funções é crucial. A simples indicação da amplitude de uma onda senoidal não é suficiente, pois o valor instantâneo da tensão é vital para requisitos de isolamento elétrico.<p/>

<LI><p align="justify"> calcular a média dessa onda não é eficaz devido à sua alternância entre positivo e negativo, resultando em uma média zero. Portanto, o valor quadrático médio (RMS) oferece uma representação precisa da corrente ou tensão fornecida à carga. <p/>

<LI><p align="justify">O RMS é calculado elevando ao quadrado a função, obtendo a média e a raiz quadrada dessa média. Isso elimina os valores negativos e fornece uma representação significativa da magnitude, como ilustrado na imagem com a linha correspondente ao valor médio da função. O RMS é essencial para determinar a quantidade correta de potência em sistemas elétricos.<p/>

<LI><p align="justify"> A seguir temos uma imagem representando a onda senoidal, a senoide elevada ao quadrado e correspondente valor RMS.<p/>

```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

amplitude = float(input("Designe a amplitude da onda: "))

t = np.arange(0, 2*pi, 0.001)
sen1 = amplitude*np.sin(t)
sen2 = amplitude*(np.sin(t))**2
Valor_Rms = amplitude*np.sqrt(1/2) + t*0

fig, senoide = plt.subplots()
senoide.plot(t, sen1, 'b', t, sen2, t, Valor_Rms)

senoide.set(xlabel='time (s)', ylabel='Amplitude',
       title= 'Senoide: $A \cdot sin(\omega \cdot t + \phi$)')

senoide.grid()
```

```python
%%manim -qm -v WARNING Senoide

class Senoide(Scene):
    def construct(self):
        self.camera.background_color=BLACK
        plane = NumberPlane(
            x_range = (0,2*PI),
            y_range = (-3,3),
            axis_config={
                "include_ticks": True,
                "include_tip": None,
            },
        )
        plane.add_coordinates().scale(0.8)

        curve = plane.plot(lambda x: 2 * np.sin(x), color=RED)
        curv2 = plane.plot(lambda x: 2 * np.sin(x)**2, color=GREEN)

        area = plane.get_area(
            curv2,
            color=(GREEN_B, GREEN_D),
            opacity=0.5,
        )

        label_center = Tex("Senoide (RMS):", color = ORANGE)
        label_center2 = Tex("A $\cdot$ sin($\omega$t + $\Phi$)")
        label_center2.next_to(label_center, RIGHT)

        labelx = Tex("Tempo (s)").scale(0.8).next_to(plane, DOWN)
        labely = plane.get_y_axis_label(Tex("Amplitude").rotate(90*DEGREES), edge = LEFT, direction = LEFT)

        Labels = VGroup(label_center, label_center2).next_to(plane,UP)
        self.add(plane, curve, curv2, Labels, area, labelx, labely)
```

<UL>

<LI><p align="justify">Como a senoide ao quadrado reside no polo positivo da amplitude, pode-se obter uma média significativa. Por conta da senoide ser simétrica temos que sua média é metade da amplitude máxima. Portanto, renormalizando o valor médio para a curva original. Para isso, já que para obtermos o valor médio elevamos a função ao quadrado, para renormalizarmos teremos que tirar a raiz quadrada da mesma, como é demonstrado nas equações seguintes:<p/>

$$
V^2_{médio} = \frac{1}{2}V_{máx}  \therefore V_{médio} = \sqrt{\frac{1}{2}} V_{máx}
$$

Logo;

$$
V_{eficaz} = \frac{1}{\sqrt{2}} \cdot V_{máx} \approx 0,707 \cdot V_{máx}
$$

Assim, temos que o valor eficaz da curva senoidal é 0,707 da amplitudade original.

## Reatâncias
---
<UL>

<LI><p align="justify">Assim como a resistência é uma propriedade de um dispositivo elétrico, em que ele dificulta o fluxo de corrente elétrica, a reatância também é uma propriedade de um dispositivo elétrico, no caso ela é responsável por influenciar no tempo relativo entre a tensão e a corrente, tendo em vista que a reatância é proporcionada de acordo com a geometria do dispositivo elétrico podendo apresentar um atraso da corrente ou um adianto da mesma a depender da frequência.<p/>

<LI><p align="justify">Com isso, pode-se dividir a reatância em duas: a indutiva, baseada na indutância e a capacitiva, baseada no capacitor. Outrossim, um dispositivo elétrico pode possuir tanto a propriedade de resistência quanto a reatância , que a junção das duas em um mesmo dispositivo, originando a impedância. Tanto a impedância, resistência e reatância são medidas em ohms ($\Omega$).<p/>

### Indutância
---
<UL>

<LI><p align="justify">O dispositivo indutivo básico é uma bobina de fio, chamada de indutor ou solenóide. Seu funcionamento se dá por conta da corrente elétrica produzir um campo magnético ao seu redor, com isso esse campo magnetico descreve um padrão circular quando ele se origina em um fio condutor de corrente. Quando o fio é enrolado gera uma amplificação do campo magnético, por conta que os loops individuais se somam. As somas dessas contribuições são especialmente grandes no centro do solenóide apontando ao longo do eixo central da bobina.<p/>

<LI><p align="justify"> Quando uma bobina é inserida em um circuito de corrente alternada (CA), a variação do campo magnético próximo a um fio condutor gera uma corrente elétrica induzida que flui através desse fio. Essa corrente induzida é diretamente proporcional à taxa de variação do campo magnético, ou seja, à frequência com que o campo magnético muda. A corrente induzida age contra a mudança no fluxo da corrente que originalmente produziu o campo magnético.<p/>

<LI><p align="justify">Portanto, um indutor atua como um dispositivo que atrasa a corrente em relação à tensão aplicada, o que é ilustrado na imagem a seguir. Esse atraso na corrente é uma característica importante dos indutores em circuitos de CA.<p/>







```python
%%manim -qm -v  WARNING Senoide

class Senoide(Scene):
    def construct(self):
        self.camera.background_color=BLACK
        plane = NumberPlane(
            x_range = (0,2*PI),
            y_range = (-3,3),
            axis_config={
                "include_ticks": True,
                "include_tip": None,
            },
        )
        plane.add_coordinates()

        curve = plane.plot(lambda x: 2 * np.sin(x), color=RED)
        curv2 = plane.plot(lambda x:  np.sin(x + PI/2), color=GREEN)

        label_center = Tex("Senoide:", color = ORANGE)
        label_center2 = Tex("A $\cdot$ sin($\omega$t + $90$)")
        label_center2.next_to(label_center, RIGHT)

        Labels = VGroup(label_center, label_center2).next_to(plane,UP)
        self.add(plane, curve, curv2, Labels)
```

<UL>

<LI><p align="justify">Quando um campo elétrico oscila, ele armazena e libera energia repetidamente, em vez de dissipá-la, sendo crucial na transferência de potência. A corrente induzida em um indutor está relacionada à rapidez com que o campo magnético muda por unidade de tempo, tornando a frequência dos ciclos relevante. Frequências mais altas dificultam a mudança no campo magnético, enquanto frequências mais baixas permitem que a corrente passe mais facilmente pelo indutor.<p/>

<LI><p align="justify">Em corrente contínua (CC), o indutor se comporta como um fio comum, pois não há variação do campo magnético devido à tensão constante. Portanto, a propriedade indutiva não afeta a corrente.<p/>

<LI><p align="justify"> O efeito do indutor em circuito CA é determinado por sua reatância, denotada por $X_{L}$, o $L$ é para identificar a reatância indutiva, a reatância indutiva pode ser encontrada pelo produto da frenquência angular e a indutância, como pode ser notado na equação a seguir: <p/>

$$
X_{L} = \omega \cdot L = 2 \cdot \pi \cdot F \cdot L
$$

<LI><p align="justify"> Com isso, a reatância não é determinada apenas pelas características intrísecas de um dispositivo. No entanto, em sistemas de potência a reatância é tratada como constante, por conta que a frequência é sempre a mesma.<p/>

### Capacitância
---
<UL>

<LI><p align="justify">Outro de tipo de reatância é a capacitiva, cujo efeito é oposto ao da reatância indutiva, o dispositivo elétrico que contém essa propriedade é o capacitor.<p/>

<LI><p align="justify">Um capacitor é um componente composto por duas placas condutoras separadas por um pequeno espaço, criando um campo elétrico entre elas. Esse campo elétrico retém o acúmulo de cargas nas placas, evitando o vazamento de carga entre elas. Capacitores são frequentemente feitos de folhas flexíveis que são dobradas ou enroladas em um recipiente, não possuindo uma forma ideal, assim como os indutores.<p/>

<LI><p align="justify">Em corrente contínua (CC), o capacitor age como uma barreira, acumulando gradualmente carga em suas placas até que elas estejam completamente carregadas, bloqueando a passagem de corrente e funcionando como um circuito aberto.<p/>

<LI><p align="justify"> Em corrente alternada (CA), os elétrons não percorrem distâncias significativas no condutor. Em vez disso, eles transferem impulsos elétricos entre si. Isso permite que os impulsos passem pela lacuna entre as placas do capacitor, desde que o fluxo de corrente continue alternando. Quando a carga se acumula nas placas, cria um campo elétrico estável, interrompendo a transmissão de impulsos elétricos.<p/>

<LI><p align="justify"> Com isso, quanto maior a frequência, mais prontamente a corrente é transmitida, pois a taxa de variação da tensão será maior. A reatância capacitiva é denotada por $X_{C}$  e é dado pelo inverso do produto da frequência com a capacitância, como pode ser visto na imagem a seguir:<p/>

$$
X_{C} = \frac{1}{\omega \cdot C} = \frac{1}{2 \cdot \pi \cdot F \cdot C}
$$

<LI><p align="justify"> Dessa forma a equação mostra que a magnitude da reatância capacitiva aumenta com a diminuição da capacitância, pois, com a diminuição da capacitância significa que as placas estão se tornando menos eficazes em suportar um campo elétrico para transmitir qualquer impulso.<p/>

<LI><p align="justify">Analogo ao indutor, o capacitor faz com que a corrente defase em relação a tensão, no entanto, o efeito é contrário, pois o indutor atrasa a corrente e o capacitor adianta a mesma, para um caso da reatância pura. Como pode ser notado na imagme a seguir, o capacitor adianta em 90° a corrente em relação a tensão. <p/>



```python
%%manim -qm -v  WARNING Senoide

class Senoide(Scene):
    def construct(self):
        self.camera.background_color=BLACK
        plane = NumberPlane(
            x_range = (0,2*PI),
            y_range = (-3,3),
            axis_config={
                "include_ticks": True,
                "include_tip": None,
            },
        )
        plane.add_coordinates()

        curve = plane.plot(lambda x: 2 * np.sin(x), color=RED)
        curv2 = plane.plot(lambda x: np.sin(x - PI/2), color=GREEN)

        label_center = Tex("Senoide:", color = ORANGE)
        label_center2 = Tex("A $\cdot$ sin($\omega$t + $90$)")
        label_center2.next_to(label_center, RIGHT)

        Labels = VGroup(label_center, label_center2).next_to(plane,UP)
        self.add(plane, curve, curv2, Labels)
```

<UL>

<LI><p align="justify">Com isso temos que o pico de corrente corresponde ao momento em que há maior variação de tensão sobre o capacitor. O capacitor também armazena  e libera energia como o indutor e sues intervalos de liberação da energia são contrarios ao do indutor. Com isso, temos que quando um indutor e um capacitor são postos em um mesmo circuito suas reatâncias se subtraem.<p/>

### Representação Complexa
---
<UL>
<LI><p align="justify"> Os números complexos são uma representação matemática que combina a resistência (parte real) e a reatância (parte imaginária) em um único número chamado impedância. A parte real corresponde a quantidades mensuráveis, enquanto a parte imaginária representa comportamento oscilatório.<p/>

<LI><p align="justify"> O número imaginário é um múltiplo da unidade imaginária $j$, onde $j$ é definido como a raiz quadrada de -1, ou seja, $j^2 = -1$. Isso introduz a ideia de oscilação na matemática. <p/>

<LI><p align="justify">Ao multiplicar por $j^2$, giramos o eixo em 180°, invertendo os números positivos e negativos. Multiplicando por $j$ novamente, giramos o eixo em 90°, introduzindo uma direção imaginária perpendicular à reta numérica real.<p/>

<LI><p align="justify">No plano complexo, números complexos são representados como pontos. Um número complexo $C = a + jb$, onde $a$ é a parte real e $b$ é a parte imaginária. O plano complexo é usado para visualizar números complexos como pontos, onde $a$ representa a posição ao longo do eixo real (Re) e $b$ ao longo do eixo imaginário (Im).<p/>

<LI><p align="justify">A representação polar utiliza vetores para descrever números complexos, onde o comprimento do vetor, ou sua magnitude, é representado como $|C|$, e o ângulo entre o vetor e o eixo real é $\theta$. Isso é útil para cálculos que envolvem ângulos e magnitude em números complexos.</p>





```python
#imagem fasor

%%manim -qm -v  WARNING Senoide

class Senoide(Scene):
    def construct(self):
        self.camera.background_color=GRAY_E
        plane = ComplexPlane(
            x_range = (-2,4),
            y_range = (-2,4),
            axis_config={
                "include_ticks": True,
                "include_tip": None,
            },
        )
        plane.add_coordinates()

        fasor = Vector([3,4], color = PURPLE, tip_shape=ArrowCircleFilledTip).shift(DL)

        self.add(plane, fasor)

```

<UL>

<LI><p align="justify">A forma demonstrada na figura acima é a representação retangular, tendo também sua representação polar, como são demontradas a seguir:</p>

$$
Retangular: C = a + jb
$$

$$
Polar: C = |C|\cdot (cos(\theta) + jsen(\theta))
$$

<LI><p align="justify">A adição e subtração de números complexos são diretas usando o formato de componente retangular, combinando suas partes reais e imaginárias. Multiplicação e divisão são mais simples na forma polar, envolvendo multiplicação de magnitudes e soma de ângulos para multiplicar, e divisão de magnitudes e subtração de ângulos para dividir.</p>

<LI><p align="justify">Embora os números complexos não tenham uma interpretação direta na física, eles seguem regras matemáticas bem definidas. Suas propriedades e regras tornam-nos ferramentas poderosas para resolver problemas em matemática, física e engenharia, especialmente em áreas que envolvem alternância, como teoria das ondas, eletricidade, magnetismo e engenharia de controle.</p>



### Impedância
---
<UL>

<LI><p align="justify">A impedância ($Z$) de um dispositivo em um circuito elétrico é uma combinação de resistência e reatância, representada como um número complexo no plano complexo. Ela é fundamental para a análise de circuitos elétricos, permitindo trabalhar com grandezas vetoriais.</p>

$$ Z = R + jX $$

<LI><p align="justify">Todos os dispositivos em um sistema elétrico têm uma impedância, mesmo que a resistência ou reatância sejam insignificantes em algumas situações. A representação polar da impedância inclui um ângulo $θ$ que reflete o deslocamento de fase entre corrente e tensão, devido à natureza da reatância. Quando a reatância é indutiva, a corrente atrasa em relação à tensão (ângulo $θ$ positivo), e quando é capacitiva, a corrente adianta (ângulo $θ$ negativo).</p>

<LI><p align="justify">Essa interpretação do ângulo de deslocamento de fase é essencial em aplicações como transmissão de sinais, controle de potência, motores elétricos e sistemas de filtragem, onde é necessário sincronizar corrente e tensão. A impedância é uma ferramenta essencial para entender e projetar circuitos elétricos e sistemas de energia.</p>

### Definição de Energia Elétrica
---
<Ul>

<LI><p align="justify">Potência é uma medida da rapidez com que a energia é usada ou produzida. É comumente expressa em Watts (W) e ajuda a quantificar o ritmo de uso ou geração de energia. A potência total usada por um dispositivo ao longo do tempo pode ser obtida multiplicando sua potência pelo tempo, o que resulta na energia total consumida, medida em Watt-hora (Wh).<p/>

<LI><p align="justify">A potência dissipada em um elemento de circuito é aquela que é convertida em calor, calculada multiplicando a resistência do elemento pelo quadrado da corrente que passa por ele, como em $P = I^2 \cdot R$. Isso é especialmente relevante em dispositivos de aquecimento elétrico.<p/>

<LI><p align="justify">Outra forma de calcular a potência é multiplicar a corrente pela tensão no elemento de circuito, usando a lei de Ohm. No entanto, ao considerar uma linha de transmissão, é importante distinguir entre a potência dissipada (determinada pela resistência) e a potência transmitida (que leva em conta a reatância da linha).<p/>

<LI><p align="justify">A reatância, causada pelas mudanças de fase na linha, complica o cálculo da potência quando multiplicamos corrente por tensão. Assim, a fórmula $P = I^2 \cdot R$ é mais precisa para calcular perdas térmicas em linhas de transmissão.<p/>

<LI><p align="justify">A potência transmitida por uma linha é calculada como $P = I \cdot V$, onde $V$ refere-se à tensão percebida pela carga conectada. Isso nos ajuda a entender como a energia é transferida e usada na eletricidade, considerando a relação entre corrente, tensão e energia. A potência é a taxa de transferência de energia elétrica, relacionando as unidades de carga, energia e tempo. Essa relação é fundamental para compreender o uso e a transferência de energia elétrica em circuitos elétricos.<p/>

### Potência Complexa
---
<UL>

<LI><p align="justify"> A fórmula $P = I \cdot V$, que é simples de aplicar em situações com tensão e corrente constantes, torna-se mais complexa quando a tensão e a corrente variam ao longo do tempo, como em sistemas de corrente alternada (CA).<p/>

<LI><p align="justify"> Em sistemas CA, a potência é tratada como uma quantidade complexa e é representada como $S = I^* \cdot V$, onde $S$ é a potência aparente. A notação com um asterisco indica o complexo conjugado da corrente $I$, exigindo a inversão do sinal da parte imaginária da corrente para cálculos.<p/>

<LI><p align="justify">Para uma compreensão conceitual mais profunda, é necessário o uso de fasores e quantidades complexas. No entanto, a relação fundamental $P = I \cdot V$ ainda é válida para uma descrição instantânea da potência, onde a potência em um determinado momento é igual ao produto da corrente e tensão naquele instante específico, considerando cada variável como uma função do tempo.<p/>

<LI><p align="justify">Em sistemas elétricos, muitas vezes, estamos interessados na potência média ao longo de intervalos de tempo maiores. Para uma carga puramente resistiva, onde a corrente e a tensão estão em fase, a potência média pode ser calculada como o produto médio da tensão e corrente eficazes. Isso nos fornece uma maneira de entender a potência média consumida ou fornecida pela carga ao longo do tempo.<p/>

<LI><p align="justify">Cargas resistivas, como aquecedores elétricos, extraem potência real dos circuitos devido à sua resistência elétrica. Por outro lado, cargas reativas, como motores elétricos ou dispositivos com indutores e capacitores, não consomem potência real, mas sim potência reativa. Isso ocorre porque a energia armazenada em um componente reativo em um momento é devolvida ao sistema em outro momento.

<LI><p align="justify">Em redes elétricas complexas, a gestão da potência reativa é essencial para garantir a eficiência e a estabilidade do sistema. As concessionárias de energia elétrica ajustam o suprimento de energia CA para equilibrar a potência reativa das cargas. Além disso, diferentes fatores de potência de cargas individuais se combinam para formar um fator de potência agregado para todo o conjunto de cargas. Esse fator agregado depende das combinações de impedâncias em série e paralelo e tende a ser mais próximo do fator de potência da carga com maior FP em uma conexão paralela.

### O significado de potência reativa
---

Um baixo fator de potência é indesejável para as concessionárias de energia devido a problemas de eficiência e custos operacionais. A energia reativa, que não realiza trabalho útil, causa perdas e requer equipamentos de maior capacidade. Isso é conhecido como "o colesterol das linhas de energia".

Para melhorar o fator de potência e reduzir perdas, as concessionárias utilizam capacitores para compensar a energia reativa, preferencialmente instalados próximos à carga.

As perdas de linha são perdas reais de potência devido ao aquecimento das linhas e não devem ser confundidas com as perdas reativas, que ocorrem devido à reatância indutiva das linhas e transformadores. A gestão da energia reativa é relevante em energia renovável e conservação de energia, pois afeta a eficiência de dispositivos como lâmpadas fluorescentes e turbinas eólicas.

Para clientes que não pagam pela energia reativa, o fator de potência pode não ser relevante, mas para as concessionárias, um baixo fator de potência resulta em custos adicionais devido a perdas e capacidade de transmissão e distribuição.

### Fasores como exponenciais
---

O fasor escrito na notação exponencial transmite todas as informações essenciais sobre a quantidade desejada, como a tensão ou a corrente, sem a necessidade de recorrer a um diagrama. Um fasor de tensão é representado como
$Ve^{j\theta}$, onde V é a magnitude da tensão e $\theta$ é o ângulo de fase da tensão. O fasor de corrente seria $Ie^{j\theta}$, com I sendo a magnitude da corrente e $\theta$ sendo o ângulo de fase da corrente. Às vezes, diferentes letras ou subscritos, como $θ_{V}$ e $θ_{I}$, são usados para evitar confusão entre os ângulos de fase.

O número $e$ é a base exponencial natural, e j é usado para representar a unidade imaginária (neste caso, em engenharia elétrica, j é usado para representar a raiz quadrada de -1). Os engenheiros frequentemente usam uma notação ainda mais abreviada que omite o $e$ e o j, assumindo que esses elementos estão implícitos. Assim, $Ve^{jθ}$ se torna $V∠θ$, onde o símbolo ∠ denota "ângulo". Isso é traduzido para um diagrama fasorial, onde V corresponde ao comprimento do fasor de tensão e θ ao ângulo que o fasor forma com o eixo horizontal.

Mas qual é o significado da exponencial nessa notação? A exponencial serve como uma abreviação para representar uma onda senoidal usando a relação conhecida como equação de Euler:

$e^{j\theta}=cos(θ)+jsin(θ)$

Essa relação é uma maneira de conectar as funções trigonométricas seno e cosseno com a notação exponencial. Ela é especialmente útil porque simplifica muitos cálculos e operações envolvendo ondas senoidais. Usando a equação de Euler, é possível expressar senoides como funções exponenciais complexas, o que facilita manipulações matemáticas e análises em circuitos elétricos.

Entender a expressão $e^{jθ}$ pode ser um conceito desafiador, mas vamos explorar um pouco mais a sua natureza para torná-la mais acessível.

Primeiro, considere a parte $e^{jθ}$ sem o j imaginário. Nesse caso, $e^θ$ simplesmente cresceria exponencialmente à medida que θ aumentasse. No entanto, o j no expoente introduz um comportamento interessante. O j causa uma "rotação" da expressão no plano complexo.

Para compreender visualmente esse comportamento, você pode experimentar inserir alguns valores de θ (que pode ser qualquer número) na expressão $e^{jθ}$ e plotar os resultados. Ao fazer isso, você verá que os resultados têm componentes oscilando entre valores positivos e negativos. A combinação de componentes reais e imaginários faz com que os pontos fiquem confinados de certa forma, mantendo uma distância constante da origem.

No entanto, essa distância constante mantida pelos pontos é uma característica importante. Ela faz com que a magnitude dos pontos (a magnitude de $e^{jθ}$) permaneça constante, formando uma flecha de comprimento invariável que gira em torno da origem. A medida que aumentamos o valor de θ no sentido anti-horário, essa flecha gira no plano complexo.

O valor de θ volta para onde começou (na direção real, com magnitude 1) conforme avançamos um ciclo completo no sentido anti-horário. Esse valor de
θ, não por coincidência, é 2π, o número de radianos em um círculo completo.



```python
#imagem, varios fasores

%%manim -qm -v  WARNING Senoide

class Senoide(Scene):
    def construct(self):
      fasor1 = Vector([1,0], color = TEAL)
      fasor2 = Vector([np.sqrt(2)/2,-np.sqrt(2)/2], color = TEAL)
      fasor3 = Vector([-1,0], color = TEAL)
      fasor4 = Vector([0,1], color = TEAL)
      fasor5 = Vector([np.sqrt(2)/2,np.sqrt(2)/2], color = TEAL)
      fasor6 = Vector([0,-1], color = TEAL)
      fasor7 = Vector([-np.sqrt(2)/2,-np.sqrt(2)/2], color = TEAL)
      fasor8 = Vector([-np.sqrt(2)/2,np.sqrt(2)/2], color = TEAL)

      circulo = Circle(color = WHITE)

      fasores = VGroup(fasor1, fasor2, fasor3, fasor4, fasor5, fasor6, fasor7, fasor8, circulo).scale(2)

      label1 = Tex(r"$e^{j2\pi}$").next_to(fasor1, RIGHT)
      label2 = Tex(r"$e^{j\frac{7\pi}{4}}$").next_to(fasor2, DR)
      label3 = Tex(r"$e^{j\pi}$").next_to(fasor3, LEFT)
      label4 = Tex(r"$e^{j\frac{\pi}{2}}$").next_to(fasor4, UP)
      label5 = Tex(r"$e^{j\frac{\pi}{4}}$").next_to(fasor5, UR)
      label6 = Tex(r"$e^{j\frac{3\pi}{2}}$").next_to(fasor6, DOWN)
      label7 = Tex(r"$e^{j\frac{5\pi}{4}}$").next_to(fasor7, DL)
      label8 = Tex(r"$e^{j\frac{3\pi}{4}}$").next_to(fasor8, UL)

      labels = VGroup(label1, label2, label3, label4, label5, label6, label7, label8)

      self.add(fasores, labels)
```

Na imagem acima temos que o número $e$ é especial, tendo que para todo $j\theta$ sendo expoente dele, é representado um fasor no angulo $\theta$ do círculo trigonométrico. Em caso de mudanças de base temos que os números não atuaram como $e$, no entanto outros números podem ser descritos em função de $e$, assim preservando o modulo do fasor e modificando o ângulo de movimentação do fasor, como pode ser visto no equacionamento a seguir:

$$
n^{j \theta} = e^{ln(n^{j \theta})}
 = e^{j \theta \cdot ln(n)}$$

Exatamente, o número $e$ é realmente um número especial. Quando o elevamos a um expoente imaginário $jθ$, ele efetivamente gira em círculos no plano complexo. Isso ocorre porque a combinação de $e^{jθ}$ representa uma quantidade que está em constante rotação enquanto mantém uma magnitude constante. É como se estivéssemos descrevendo algo que oscila ou gira com uma amplitude constante ao longo do tempo. Isso é especialmente útil para representar quantidades senoidais, como tensões e correntes alternadas, de uma maneira compacta e matematicamente conveniente.

### Operações com fasores
---

A notação fasorial não menciona explicitamente a frequência das ondas, porque se pressupõe que estamos operando em uma situação onde a frequência é conhecida e a mesma para todas as grandezas que variam senoidalmente. Em vez disso, os fasores enfatizam o ângulo de fase de cada quantidade ou o tempo preciso em relação a uma referência. Para estabelecer essa referência, geralmente definimos o ângulo de fase zero para uma onda de tensão específica.

Quando representamos um fasor graficamente como uma seta ou vetor, a magnitude do fasor é o comprimento da seta e o ângulo de fase é o ângulo entre a seta e um eixo de referência (geralmente o eixo x). A notação fasorial completa inclui as coordenadas de magnitude e ângulo lado a lado, normalmente indicadas em negrito, como $V = V_{rms} \angle \theta$ e $I = I_{rms} \angle \theta$, onde
θ representa o ângulo de fase.

A notação fasorial torna possível expressar relações complexas de forma concisa. A Lei de Ohm, por exemplo, pode ser expressa como $V=IZ$, onde
Z é a impedância. Quando multiplicamos ou dividimos fasores, multiplicamos ou dividimos suas magnitudes e somamos ou subtraímos seus ângulos para obter os resultados correspondentes.

O uso da notação fasorial simplifica o cálculo e a compreensão das grandezas senoidais em sistemas elétricos, permitindo-nos focar nas relações entre as magnitudes e os ângulos de fase das grandezas, sem a necessidade de se referir explicitamente ao tempo ou à frequência. Isso torna a análise de circuitos CA mais intuitiva e eficiente.

O processo de adição e subtração de ângulos de fase funciona porque eles realmente representam expoentes da mesma base, e. Assim, se fôssemos reescrever a equação anterior à mão, seria:

$$V \cdot e^{\theta_{V}} = I \cdot e^{\theta_{I}} \cdot Z \cdot e^{\theta_{Z}}$$
$$V \cdot e^{\theta_{V}} = I \cdot Z \cdot e^{(\theta_{I} + \theta_{Z})}$$

O procedimento de adicionar ângulos em um diagrama fasorial é simples de executar graficamente. Para multiplicar os fasores I e Z, por exemplo, você desenha um novo vetor em um ângulo com o eixo x, correspondente à soma dos ângulos dos dois fasores, e com um comprimento correspondente ao produto dos comprimentos dos fasores.

Porém, é importante notar que o comprimento do novo vetor (resultante) não pode ser diretamente comparado com outros fasores que representam unidades diferentes. Se você estiver somando dois ou mais fasores, eles devem representar quantidades da mesma dimensão e serem desenhados na mesma escala para que a soma faça sentido.

Considere um circuito CA, por exemplo. No diagrama fasorial, você pode representar várias coisas sobre o circuito. Isso é feito desenhando os fasores de tensão lado a lado, adicionando-os como as quedas de tensão são adicionadas em série no circuito. O resultado final deve corresponder à queda de tensão medida no circuito, mesmo que as posições dos fasores no gráfico sejam arbitrárias.

Quando representamos os fasores graficamente e projetamos suas projeções no eixo horizontal (real), eles geram ondas senoidais para as respectivas grandezas, como as tensões $IR$ e $IX$. O diagrama fasorial oferece uma explicação visual de como o ângulo de fase ou fator de potência de um circuito é determinado pela combinação desses elementos. A representação fasorial da potência complexa S é obtida através do produto vetorial dos fasores de corrente e tensão:

$S = I* \cdot V$

O asterisco denota o conjugado complexo da corrente I, o que significa inverter o sinal do componente imaginário para fins de cálculo. Essa inversão está relacionada com a convenção de direção positiva de ângulos na análise de circuitos. A potência resultante tem um ângulo de fase positivo, que é o mesmo ângulo pelo qual a corrente está atrasada em relação à tensão.

A potência complexa S, a potência ativa (real) P e a potência reativa (reativa) Q têm a mesma dimensão física (energia por tempo), embora sejam identificadas com diferentes unidades (VA, watts e VARs). Essas grandezas podem ser representadas em um diagrama fasorial, combinando-as de forma útil para analisar as características da potência em um circuito CA.
